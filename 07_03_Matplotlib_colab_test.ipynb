2. Plot 만들기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y) # 선 그래프
plt.show()
```

3. 여러 개의 점

```python
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y, 'o') # 점만 표시
plt.show()
```

4. 마커(Markers)

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='*') # 별 모양 마커
plt.show()
```

5. 선(Line)

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle='--') # 파선
plt.plot(ypoints, ls=':') # 점선 축약
plt.show()

```

6. 색상(Color)

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color='hotpink') # 문자열
plt.plot(ypoints, c='#4CAF50') # HEX 코드
plt.show()

```

7. 선 너비(Width)

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth=5)
plt.show()
```

8. 여러 선 그리기

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, label="Line 1")
plt.plot(y2, label="Line 2")
plt.legend()
plt.show()
```

9. 라벨과 제목(Label, Title)

```python
x = np.array([80, 85, 90])
y = np.array([240, 250, 260])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Pulse")
plt.ylabel("Calories")
plt.show()
```

10. 그리드(Grid)

```python
x = np.array([80, 85, 90])
y = np.array([240, 250, 260])

plt.plot(x, y)
plt.grid()
plt.show()
```

11. 서브플롯(Subplots)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)

plt.show()
```
