6번 + 예시 실행 코드 포함

```python
# 장애물 감지 함수 정의
def detect_obstacles(sensors, safe_distance):
    print("함수 시작")
    print("입력 데이터:", sensors, safe_distance)

    obstacles = []

    for direction, distance in sensors.items():
        print("현재 방향과 거리 체크:", direction, distance)

        if distance < safe_distance:
            obstacles.append(direction)
            print("위험! 리스트에 추가:", direction)
        else:
            print("안전:", direction)

    print("장애물 감지 종료, 최종 obstacles 리스트:", obstacles)
    return obstacles


# 예시 실행 코드
if __name__ == "__main__":
    # 센서 거리 데이터
    sensors = {
        "앞": 3.0,
        "왼쪽": 15.0,
        "오른쪽": 7.5
    }

    # 안전 거리 기준
    safe_distance = 10.0

    # 함수 실행
    result = detect_obstacles(sensors, safe_distance)

    # 결과 출력
    print("\n 최종 감지된 장애물 방향:", result)
```

```python
def decide_action(obstacles):
    # 장애물 정보 입력받음
    if len(obstacles) == 0:
        print("출력: 장애물 없음")
        return "직진"

    if "앞" in obstacles:
        print("출력: 앞에 장애물 있음, 좌우 확인")

        if "왼쪽" not in obstacles:
            print("출력: 왼쪽 안전")
            return "좌회전"
        elif "오른쪽" not in obstacles:
            print("출력: 오른쪽 안전")
            return "우회전"
        else:
            print("출력: 모든 방향 위험")
            return "정지"

    elif "왼쪽" in obstacles and "오른쪽" not in obstacles:
        print("출력: 앞은 안전, 왼쪽 위험")
        return "주의해서 직진"

    print("출력: 장애물 없음")
    return "직진"
```

7번 + 예시 실행 코드 포함

```python
def decide_action(obstacles):
    if len(obstacles) == 0:
        print("출력: 장애물 없음")
        return "직진"

    if "앞" in obstacles:
        print("출력: 앞에 장애물 있음, 좌우 확인")

        if "왼쪽" not in obstacles:
            print("출력: 왼쪽 안전")
            return "좌회전"
        elif "오른쪽" not in obstacles:
            print("출력: 오른쪽 안전")
            return "우회전"
        else:
            print("출력: 모든 방향 위험")
            return "정지"

    elif "왼쪽" in obstacles and "오른쪽" not in obstacles:
        print("출력: 앞은 안전, 왼쪽 위험")
        return "주의해서 직진"

    print("출력: 장애물 없음")
    return "직진"


# 예시 실행
if __name__ == "__main__":
    # 테스트 케이스 1: 장애물 없음
    print("\n[예시 1] 장애물 없음")
    obs1 = []
    action1 = decide_action(obs1)
    print(" 행동 결정:", action1)

    # 테스트 케이스 2: 앞에만 장애물
    print("\n[예시 2] 앞에만 장애물")
    obs2 = ["앞"]
    action2 = decide_action(obs2)
    print(" 행동 결정:", action2)

    # 테스트 케이스 3: 앞 + 왼쪽에 장애물
    print("\n[예시 3] 앞 + 왼쪽 장애물")
    obs3 = ["앞", "왼쪽"]
    action3 = decide_action(obs3)
    print(" 행동 결정:", action3)

    # 테스트 케이스 4: 앞 + 좌우 모두 장애물
    print("\n[예시 4] 모든 방향 장애물")
    obs4 = ["앞", "왼쪽", "오른쪽"]
    action4 = decide_action(obs4)
    print(" 행동 결정:", action4)

    # 테스트 케이스 5: 앞은 안전, 왼쪽만 장애물
    print("\n[예시 5] 왼쪽만 장애물")
    obs5 = ["왼쪽"]
    action5 = decide_action(obs5)
    print(" 행동 결정:", action5)
```
